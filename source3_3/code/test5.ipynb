```python
def des_sort(apper_dict):
    appear_list = sorted(apper_dict.items(), key= lambda word : word[1], reverse=True)
    return appear_list
```

```python
def mapping(line):
    appear_dict = {}
    line = line.split('\t')
    for word in line[1:]:
        if word not in appear_dict.keys():
            appear_dict[word] = 1
        else:
            tmp = appear_dict.get(word) + 1
            appear_dict[word] = tmp
    appear_list = des_sort(appear_dict)
    return appear_list
```

```python
def excute(i):
    dic = {}
    each_commit = []
    with open('../../../../sample_data/data/camel_snippet_each_commit/commit_snippet_camel_'+str(i)+'.txt') as f:
        while True:
            appear_dict = {}
            line = f.readline()
            if line == '':
                break
            tmp = line.split('\t')
            # if tmp[1].strip() != '':
            #     dic[tmp[0]] = len(tmp[1:])
            dic[tmp[0]] = len(tmp[1:])
            appear_list = mapping(line)
            each_commit.append(len(appear_list))
            

    # commit_list = []
    # for key in dic.keys():
    #     commit_list.append(key)
        
    length_df = pd.DataFrame(
        data= {'commit_hash':dic.keys(),
            'snippet_length':dic.values(),
            'word_kinds':each_commit,
            }
    )
    
    length_df.to_csv('../label/commit_camel_'+str(i)+'.csv')
```
