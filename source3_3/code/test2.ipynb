```python
ratio_by_change_type = df.groupby(['commit_hash', 'contains_bug'])['change_type'].value_counts(normalize=True) * 100
```

```python
df2 = ratio_by_change_type.reset_index(level=['contains_bug','change_type'])
df2
```

```python
df3 = df2[df2['change_type'] == 'M']
df3
```

```python
df3['contains_bug'].value_counts() / df3.shape[0]
```

```python
df3 = df3.reset_index(level='change_type')
```

```python
df3
```

```python
df4 = df3[df3['change_type'] == 'A']
```

```python
df4
```

```python
import matplotlib.pyplot as plt
bins = [0, 50, 100]
labels = ['0-50','50-100']  # 各範囲のラベル
df4['proportion_range'] = pd.cut(df4['proportion'], bins=bins, labels=labels)
change_type_counts = df4.groupby(['proportion_range', 'change_type']).size().unstack()
change_type_counts.plot(kind='bar', stacked=True)
plt.xlabel('proportion_count')
plt.ylabel('Count')
plt.title('propotion count by A')
plt.legend(title='Change Type')

plt.show()
```
